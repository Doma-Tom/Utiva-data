```python
# THIS IS AN AFRICAN DATA FROM A BREWERY COMPANY.
#  I ANSWERED THE QUESTIONS AND WRAPPED THEM IN A FUNCTION.



import time
import numpy as np
import pandas as pd

international_brewery = {'brewery_data': "brewery_data.csv"}

def get_filter():
    data = input("this is the data: ")
    city = international_brewery[data]
    print("-"*60)
    return city

def load_data(city):
    df = pd.read_csv(city)
    print(df.head(5))
    return df
    print("-"*60)


def col_lums(df):
    columns = df.columns
    print(columns)
    print("-"*60)
    
def sum_profit(df):
    sumprofit = df["PROFIT"].sum()
    print(f"sum of profit is {sumprofit}")

    
def all_countries(df):
    countries = df["COUNTRIES"].unique()
    print(f"all countries present are {countries}")
    print("-"*60)
    
# Within the space of the last three years, what was the profit worth of the breweries,
# inclusive of the anglophone and the francophone territories?

def francophone(df):
    french_speaking = df[(df.COUNTRIES =="Benin")|(df.COUNTRIES =="Senegal")|(df.COUNTRIES =="Togo")]
    sum_profit_french = french_speaking["PROFIT"].sum()
    print(french_speaking.head(5))
    print(f" sum of profit for francophone countries is {sum_profit_french}")

def anglophone(df):
    english_speaking = df[(df.COUNTRIES =="Nigeria") | (df.COUNTRIES == "Ghana")]
    sum_profit_english = english_speaking["PROFIT"].sum()
    print(english_speaking.head(5))
    print(f" sum of profit for anglophone countries is {sum_profit_english}")
    print("-"*60)
    
# Compare the total profit between these two territories in order for the territory manager,
# Mr. Stone made a strategic decision that will aid profit maximization in 2020

def country_profit_2019(df):
    country_profit = df.groupby("COUNTRIES")["PROFIT"].sum().sort_values(ascending=False)
    print(country_profit)
    print("-"*60)
    
# Country that generated the highest profit in 2019

def Year_high_profit(df):
    year_profit = df.groupby("YEARS")["PROFIT"].sum().sort_values(ascending=False)
    print(year_profit)
    print("-"*60)
    
# Which month in the three years was the least profit generated?

def month_profit_3years(df):
    month_profit_list =df.groupby("MONTHS")["PROFIT"].sum().sort_values(ascending=True)
    data_f = month_profit_list.reset_index() 
    mon_min = data_f["PROFIT"]=="MONTHS"
    mon =data_f[mon_min]
    k = mon.min()
    print(data_f)
    print(k)
    print("-"*60)
    
# What was the minimum profit in the month of December 2018?

def mon_dec_2018(df):
    dec = df["MONTHS"]=="December"
    dec_year =df[dec]
    year= dec_year[(dec_year.YEARS== 2018)]
    dec_min = year["PROFIT"].min()
    print(f"MINIMUM PROFIT IN DECEMBER IS {dec_min}")
    print("-"*60)
    
# Compare the profit in percentage for each of the month in 2019

def profit_Percent(df):
    profit = df[(df.YEARS == 2019)]
    suma = profit["PROFIT"].sum()
    arr_profit = np.array(profit["PROFIT"])
    for i in arr_profit:
            percent =((i/suma) * 100)               
            print(f"the profit for {i} is : {round(percent,2)}%")
            print(f"sum of profit for year 2019 is {suma}")
    print("-"*60)
    
#   Help him find the year with the highest profit.

def highest_profit(df):
    yr_1= df[(df.YEARS == 2017)]
    yr_2017 = yr_1["PROFIT"].sum()
    yr_2= df[(df.YEARS ==2018)]
    yr_2018 =yr_2["PROFIT"].sum()
    yr_3= df[(df.YEARS == 2019)]
    yr_2019 = yr_3["PROFIT"].sum()
    if yr_2019 > yr_2018 and yr_2019 >yr_2017:
        print(f"the year with the highest profit is 2019 with value:{yr_2019}")
    elif yr_2018>yr_2019 and yr_2018 > yr_2017:
        print(f"the year with the highest profit is 2018 with value :{yr_2018}")
    elif yr_2017 > yr_2019 and yr_2017 > yr_2018:
        print(f"the year with the highest profit is 2017 with value :{yr_2017}")
    else:
        pass
    print("-"*60)
 
    
#     profit_year = [("y_2017",yr_2017),("y_2018",yr_2018),("y_2019",yr_2019)]
# def high_profit(profit_year):
#         profit_max = 0
#         year =""
#         for key,item in profit_year:
#             if item > 0:
#                 profit_max =item
#                 year = key
#             else:
#                 pass
#             print(year,profit_max)
            
def senegal(df):
    df_sen =df[(df.COUNTRIES =="Senegal")]
    senel_brand = df_sen.groupby("BRANDS")["PROFIT"].sum().sort_values(ascending=False).head(1)
    print(f"In senegal, the brand with highest profit is {senel_brand}")
    print("-"*60)
    
#BRAND ANALYSIS
                
#1. Within the last two years, the brand manager wants to know the top three brands
# consumed in the francophone countries.

def brand(df):
    year = df[(df.YEARS==2019) |(df.YEARS==2018)]
    franco =year[(year.COUNTRIES =="Senegal")|(year.COUNTRIES=="Benin")|(year.COUNTRIES =="Togo")].head(3)
    top_brand = franco.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False).head(3)
    print(f"the top 3 brands is {top_brand}")
    
# Top 2 choices of brands by Ghana
def ghana_choice(df):
    counto=df[(df.COUNTRIES =="Ghana")]
    Gh_choice=counto.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False).head(2)
    print(f"the top 2 choices by Ghana is {Gh_choices}")
    
    
#Find out the details of beers consumed in the past three years in the most oil reached
#country in West Africa

def nigeria(df):
    oil_rich= df[(df.COUNTRIES=="Nigeria")]
    oil_rich_beer =oil_rich[(oil_rich.BRANDS=="trophy")|(oil_rich.BRANDS=="budweiser")|(oil_rich.BRANDS=="castle lite")|(oil_rich.BRANDS=="eagle lager")|(oil_rich.BRANDS =="hero")]
    print(oil_rich_beer)
    print("-"*60)
    
# Favorites malt brand in Anglophone region between 2018 and 2019
def malt_brand(df):
    english_speaking = df[(df.COUNTRIES =="Nigeria") | (df.COUNTRIES == "Ghana")]
    malt=english_speaking[(english_speaking.YEARS ==2019)|(english_speaking.YEARS==2018)]
    fav = malt[(malt.BRANDS=="grand malt")|(malt.BRANDS=="beta malt")]
    fav_malt = fav.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False)
    print(fav_malt)
    print("-"*60)
    
# # Which brands sold the highest in 2019 in Nigeria?
def sold_highest(df):
    oil_rich = df[(df.COUNTRIES=="Nigeria")]
    sold=oil_rich[(oil_rich.YEARS==2019)]
    so_high= sold.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False).reset_index()
    print(so_high)
    print("-"*60)
    
# Favorites brand in South_South region in Nigeria
def south_south(df):
    oil_rich =df[(df.COUNTRIES=="Nigeria")]
    fav_so= oil_rich[oil_rich["REGION "] =="southsouth"]
    so_so = fav_so.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False).reset_index()
    print(so_so)
    print("-"*60)
    
#  Bear consumption in Nigeria
def beer_Nig(df):
    oil_rich =df[(df.COUNTRIES=="Nigeria")]
    oil_rich_beer =oil_rich[(oil_rich.BRANDS=="trophy")|(oil_rich.BRANDS=="budweiser")|(oil_rich.BRANDS=="castle lite")|(oil_rich.BRANDS=="eagle lager")|(oil_rich.BRANDS =="hero")]
    beer =oil_rich_beer.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False).reset_index()
    print(beer) 
    print("-"*60)
    
# Level of consumption of Budweiser in the regions in Nigeria
def bud_nig(df):
    oil_rich =df[(df.COUNTRIES=="Nigeria")]
    bud =oil_rich[(oil_rich["BRANDS"]=="budweiser")]
    budwei =bud.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False)
    print(f"budweiser consumption in Nigeria is\n {budwei}")
    print("-"*60)
    
    
    
# Level of consumption of Budweiser in the regions in Nigeria in 2019 (Decision on Promo)
def bud_reg_nig(df):
    oil_rich =df[(df.COUNTRIES=="Nigeria")]
    bud=oil_rich[(oil_rich["BRANDS"] =="budweiser")]
    bud_oil_rich=bud[(bud["YEARS"]==2019)]
    bud1=bud_oil_rich.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False)
    print(f"budweiser consumption in Nigeria in 2019 is\n {bud1}")
    print("-"*60)
    
    
# Country with the highest consumption of beer
def country_beer(df):
    df_beer=df[(df.BRANDS=="trophy")|(df.BRANDS=="budweiser")|(df.BRANDS=="castle lite")|(df.BRANDS=="eagle lager")|(df.BRANDS =="hero")]
    county_beer=df.groupby("BRANDS")["QUANTITY"].sum().sort_values(ascending=False).head(1).reset_index()
    print(county_beer)
    print("-"*60)

# Highest sales personnel of Budweiser in Senegal
def Sales_rep(df):
    county= df[(df.COUNTRIES =="Senegal")]
    county_bud=county[(county.BRANDS=="budweiser")]
    high_county_bud= county_bud.groupby("SALES_REP")["QUANTITY"].sum().sort_values(ascending=False).head(1).reset_index()
    print(f"The highest sale_rep for budweiser is \n{high_county_bud}")
                                                                                                                         
# Country with the highest profit of the fourth quarter in 2019

def profit_quarter(df):
    mon= df[(df["YEARS"]==2019)]
    mon_pro= mon[(mon["MONTHS"]=="October") | (mon["MONTHS"]=="November")| (mon["MONTHS"]=="December")]
    mon_pro_yr = mon_pro.groupby("COUNTRIES")["PROFIT"].sum().sort_values(ascending=False)
    print(f"country with highest profit in fourth quarter is \n{mon_pro_yr}")  
      
    
    
    
def main():
    p = get_filter()
    data_frame =load_data(p)
    colum = col_lums(data_frame)
    profit =sum_profit(data_frame)
    countries =all_countries(data_frame)
    french_speaking = francophone(data_frame)
    english_speaking =anglophone(data_frame)
    country_profit =country_profit_2019(data_frame)
    year_profit = Year_high_profit(data_frame)
    month_profit = month_profit_3years(data_frame)
    mon_2018 = mon_dec_2018(data_frame)
    sum_profit_2019yr =profit_Percent(data_frame)
    high_profit =highest_profit(data_frame)
    sen =senegal(data_frame)
    top= brand(data_frame)
    Naija=nigeria(data_frame)
    malt=malt_brand(data_frame)
    so_hi= sold_highest(data_frame)
    so_fine=south_south(data_frame)
    beer_con=beer_Nig(data_frame)
    budweis=bud_nig(data_frame)
    bud_wei_2019=bud_reg_nig(data_frame)
    count=country_beer(data_frame)
    sales=Sales_rep(data_frame)
    pr_quart=profit_quarter(data_frame)
    
    
    print("-"*60)
    
    
if __name__ == "__main__":
    main()
    
```
